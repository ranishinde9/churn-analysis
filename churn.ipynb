```python
import pandas as pd

# Load the dataset
file_path = "Churn_Modelling.csv"  # Update path if needed
df = pd.read_csv(file_path)

# Calculate the mean separately for Exited = 0 and Exited = 1 (numeric columns only)
mean_0 = df[df["Exited"] == 0].mean(numeric_only=True)
mean_1 = df[df["Exited"] == 1].mean(numeric_only=True)

# Print the results
print("Mean values for customers who did NOT exit (Exited = 0):")
print(mean_0)

print("\nMean values for customers who exited (Exited = 1):")
print(mean_1)
```

```python
df = df.drop(columns=["RowNumber", "Surname", "CustomerId"])
```

```python
df.info()
```

```python
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
df.Geography.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder()
geo_encoder=onehot_encoder_geo.fit_transform(df[['Geography']]).toarray()
geo_encoder
```

```python
df.info()
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
df.head()
```

```python
with open ('label_encoder_gender.pkl','wb')as file:
    pickle.dump(label_encoder_gender,file)
with open('onehot_encoder_geo.pkl','wb')as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
x =df.drop(['Exited'],axis = 1)
y = df.Exited
print(f'x = {x.shape},y ={y.shape}')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[early_stopping_callback]
)
```

```python
model.save('model.h5')
```
